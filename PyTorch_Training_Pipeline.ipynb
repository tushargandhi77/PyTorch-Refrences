```python
import numpy  as np 
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
df = pd.read_csv('data/breast_cancer_data.csv')
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
df.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

```python
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
y_train
```

### Label Encoding

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

#### numpy to tensor

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

## Definition of Model

```python
class MySimpleNN():
    def __init__(self,X):
        self.weights = torch.rand(X.shape[1],1,dtype=float,requires_grad=True)
        self.bias = torch.zeros(1,dtype=float,requires_grad=True)

    def forward(self,X):
        z = torch.matmul(X,self.weights) + self.bias
        y_pred = torch.sigmoid(z)
        return y_pred

    def loss_function(self,y_pred,y):
        epsilon = 1e-7
        y_pred = torch.clamp(y_pred,epsilon,1-epsilon)

        loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1-y_pred)).mean()
        return loss
```

### Important parameters

```python
learning_rate = 0.1
epochs = 25
```

## Training Pipelines

```python
# create model
model = MySimpleNN(X_train_tensor)

# define loop

for epoch in range(epochs):

    # forward pass
    y_pred = model.forward(X_train_tensor)


    # loss calculate
    loss = model.loss_function(y_pred,y_train_tensor)

    # backward pass
    loss.backward()

    # parameter update
    with torch.no_grad():
        model.weights -= learning_rate * model.weights.grad
        model.bias -= model.bias - learning_rate * model.bias.grad

    # zero grad
    model.weights.grad.zero_()
    model.bias.grad.zero_()

    # print loss in each epochs
    print(f"Epoch: {epoch+1}, Loss: {loss.item()}")
```

## Model Evaluation

```python
with torch.no_grad():
    y_pred = model.forward(X_test_tensor)
    y_pred = (y_pred > 0.7).float()
    accuracy = (y_pred == y_test_tensor).float().mean()
    print(f"Accuracy Score:-> {accuracy}")
```
